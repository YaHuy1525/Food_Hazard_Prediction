```python
import matplotlib.pyplot as plt
import numpy as np

# Data points
x_points = [1, 2, 4, 5, 7]
y_points = [3, 1, 5, 7, 6]

# Least squares line: y = 1.333 + 0.807x
x_line = np.array([0, 10])  # x range from 0 to 10
y_line = 1.333 + 0.807 * x_line

# Create the plot
plt.figure(figsize=(10, 8))
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)

# Plot the data points and regression line
plt.scatter(x_points, y_points, color='blue', s=80, zorder=5, label='Data points')
plt.plot(x_line, y_line, color='red', linewidth=2, label='y = 1.333 + 0.807x')

# Add labels and title
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Least Squares Approximating Line', fontsize=14)
plt.legend(fontsize=12)

# Set axis limits
plt.xlim(0, 10)
plt.ylim(0, 10)

# Add the equation as text on the plot
plt.text(2, 8, 'y = 1.333 + 0.807x', fontsize=12, color='red', 
         bbox=dict(facecolor='white', alpha=0.8))

# Show the plot
plt.tight_layout()
plt.show()

# Optional: Save the figure
# plt.savefig('least_squares_plot.png', dpi=300, bbox_inches='tight')
```
